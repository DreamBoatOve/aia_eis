```python
import os
folder_path = '../../datasets/goa_datasets/simulated'
def get_ecm_num_str(ecm_num):
    ecm_num_str = ''
    if ecm_num < 10:
        ecm_num_str += '00' + str(ecm_num)
    elif ecm_num < 100:
        ecm_num_str += '0' + str(ecm_num)
    else:
        return str(ecm_num)
    return ecm_num_str
```

### Evolution-based

#### CGA

```python
# Continuous Genetic Algorithm
iter_time = 5000
entity_num = 50
goa = CGA_EIS(exp_data_dict = sim_ecm_1, iter_num = iter_time, entity_num = entity_num)
cur_best_entity_list, global_best_entity_list, iter, chi_squared = goa.search()

print('Real Elements', sim_ecm_1['para'])
print('Para ranges', sim_ecm_1['limit'])
print('Iteration {0}\nEntity {1}\nFitting '.format(iter, entity_num), global_best_entity_list[-1].x_list)

x_lists_list = [entity.x_list for entity in global_best_entity_list]
rel_std_err_list = goa_rel_std_err(x_lists_list)
print('Rel. Std. Err:', rel_std_err_list)
print('Chi-Squared', chi_squared)

print('Best Fitness:', global_best_entity_list[-1].fitness)
fitness_list = [entity.fitness for entity in global_best_entity_list]
fitness_v_iter(fitness_list, entity_num=entity_num, alg_name='CGA')
```

```python
from circuits.circuit_pack import RaQRb

import math
def fre_generator(f_start, f_end, pts_decade):
    """
    :param
        f_start:
            int
            High frequency
            Ex: 7, == 1e7
        f_end:
            int
            Low frequency
            Ex: 7, == 1e-7
        pts_decade:
    :return:
    """
    points_num = (f_start - f_end) * pts_decade
    fre_list = [10 ** (f_start - i * (f_start - f_end) / points_num) for i in range(points_num)]
    w_list = [2 * math.pi * f for f in fre_list]
    return fre_list, w_list
# Frequency range 1e5 ~ 1e3, 10 points per decade
fre_list, w_list = fre_generator(f_start = 5, f_end = -3, pts_decade = 10)

"""
    Real Elements [0.02, 0.00018, 0.8, 53000.0]
    Para ranges [[1e-15, 2], [1e-15, 0.01], [0.5, 1], [1e-15, 1000000.0]]
    Iteration 1001
    Entity 20
    Fitting  [0.02243501099790816, 0.0001798817474450777, 0.8001891949019146, 53021.710813048376]
    Rel. Std. Err: [57.16911125883857, 19.80979457922295, 1.2189778823811306, 70.4321124995712]
    Chi-Squared 992.0216961797092
    Best Fitness: 0.0009920216961797096
"""
z_raw_list = sim_ecm_1['z_sim']
z_pre_list = [RaQRb(w, R0=0.02243501099790816, Q0_pair=[ 0.0001798817474450777, 0.8001891949019146], R1=53021.710813048376)\
              for w in w_list]
```

<font size=5>Calculation of Chi-Square in PyEis</font>

```python
def cal_Chi_Square_PyEis(z_raw_complex_list, z_pre_complex_list):
    # Chi-Squared
    cs = 0.0
    for z_raw, z_pre in zip(z_raw_complex_list, z_pre_complex_list):
        d_square = (z_raw.real - z_pre.real) ** 2 + (z_raw.imag - z_pre.imag) ** 2
        # mw, Modulus Weighting
        mw = math.sqrt(z_raw.real ** 2 + z_raw.imag ** 2)
        cs += d_square / mw
    return cs
cs = cal_Chi_Square_PyEis(z_raw_list, z_pre_list)
print('PyEis', cs)
```

<font size=5>Calculation of Chi-Square in Mine colde-V1</font>

```python
def cal_Chi_Square_ove(z_raw_complex_list, z_pre_complex_list):
    # Chi-Squared
    cs = 0.0
    for z_raw, z_pre in zip(z_raw_complex_list, z_pre_complex_list):
        d_square = (z_raw.real - z_pre.real) ** 2 + (z_raw.imag - z_pre.imag) ** 2
        # mw, Modulus Weighting
        sigma = 0.001
        mw = sigma * math.sqrt(z_raw.real ** 2 + z_raw.imag ** 2)
        cs += d_square / (mw ** 2)
    return cs
cs = cal_Chi_Square_ove(z_raw_list, z_pre_list)
print('ove', cs)
```

<font size=5>Calculation of Chi-Square in Mine colde-V2</font>已经改成和下方的ZSimpWin一致

<font size=5>Calculation of Chi-Square in ZSimpWin</font>

```python
def cal_Chi_Square_ZSimpWin(z_raw_complex_list, z_pre_complex_list):
    # Chi-Squared
    cs = 0.0
    for z_raw, z_pre in zip(z_raw_complex_list, z_pre_complex_list):
        d_square = (z_raw.real - z_pre.real) ** 2 + (z_raw.imag - z_pre.imag) ** 2
        # mw, Modulus Weighting
        mw_square = z_raw.real ** 2 + z_raw.imag ** 2
        cs += d_square / mw_square
    return cs
cs = cal_Chi_Square_ZSimpWin(z_raw_list, z_pre_list)
print('ZSimpWin', cs)
```
