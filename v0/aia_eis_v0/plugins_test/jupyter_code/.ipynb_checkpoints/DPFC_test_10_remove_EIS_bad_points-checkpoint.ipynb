# Remove Bad Points in EIS
<p>GOA直接对EIS拟合的结果不如赖的手动拟合结果好，可能的因素是在手动拟合的过程中可以删除一些坏点来提高拟合精度，而GOA直接对原始数据进行拟合，GOA可能受到坏点的影响</p>
<p>在此尝试找到赖的ECM2和ECM9中的一些案例，通过删除一个或多个坏点后，期望GOA的再次拟合精度能提升</p>
<p></p>

```python
def nyquist_plot(z_list, x_lim=[0, 50000], y_lim=[0, 50000], grid_flag=False, plot_label=''):
    z_real_list = [z.real for z in z_list]
    z_imag_list = [z.imag for z in z_list]
    z_inv_imag_list = [-z_imag for z_imag in z_imag_list]

    fig = plt.figure()
    plt.plot(z_real_list, z_inv_imag_list, 'o--', label=plot_label)

    plt.xlim(x_lim)
    plt.ylim(y_lim)
    plt.xlabel('Z$_{real}$ [$\Omega$]')
    plt.ylabel('-Z$_{imag}$ [$\Omega$]')
    plt.grid(grid_flag)
    plt.legend()
    plt.show()
```

```python
def read_removed_eis_file(fn, fp='./dpfc_files/tests_10/files'):
    fre_list = []
    normed_z_list = []
    area = 1.01 * 1e-6
    with open(os.path.join(fp, fn), 'r', encoding='utf-8') as f:
        for line in f.readlines()[3:]:
            fre, zr, zi = [float(i) for i in line.strip().split(',')]
            normed_z = zr * area + 1j * zi * area
            fre_list.append(fre)
            normed_z_list.append(normed_z)
    return fre_list, normed_z_list
```

# Examples of ECM2

```python
def get_ECM2_para(para_list):
    R0 = para_list[0]
    Q0_pair = para_list[1:3]
    R1 = para_list[3]
    Q1_pair = para_list[4:6]
    R2 = para_list[-1]
    return R0, Q0_pair, R1, Q1_pair, R2
```
