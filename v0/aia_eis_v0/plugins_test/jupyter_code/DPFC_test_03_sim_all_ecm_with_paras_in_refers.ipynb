# 初始化电路元件，频率生成器 及绘图函数

## 初始化电路元件

```python
def ele_C(w, C):
    # w: Angular frequency [1/s]
    # C: F
    return 1 / (1j * w * C)

def ele_L(w, L):
    # w: Angular frequency [1/s]
    # L: Inductance [ohm * s]
    return 1j * w * L

def ele_Warburg(w, sigma):
    # w: Angular frequency [1/s]
    # sigma: warburg coefficient
    return sigma * (w ** 0.5) *(1 - 1j)

def ele_Q(w, q, n):
    """
    :param
        w: Angular frequency [1/s]
        q: CPE coefficient, Constant phase element [s^n/ohm]
        n: Constant phase elelment exponent [-]
    :return:
    """
    z = 1 / (q * ((1j * w) ** n))
    return z

C = ele_C
L = ele_L
WB = ele_Warburg
Q = ele_Q
```

## 初始化频率生成器

```python
import math
def fre_generator(f_start, f_end, pts_decade):
    """
    :param
        f_start:
            int
            High frequency
            Ex: 7, == 1e7
        f_end:
            int
            Low frequency
            Ex: 7, == 1e-7
        pts_decade:
    :return:
    """
    points_num = (f_start - f_end) * pts_decade
    fre_list = [10 ** (f_start - i * (f_start - f_end) / points_num) for i in range(points_num)]
    w_list = [2 * math.pi * f for f in fre_list]
    return fre_list, w_list

fre_list, w_list = fre_generator(f_start = 7, f_end = -6, pts_decade = 10)
```

## 初始化绘图函数

```python
import math
import matplotlib.pyplot as plt

def nyquist_plot(z_list, x_lim=[0, 50000], y_lim=[0, 50000], plot_label=''):
    z_real_list = [z.real for z in z_list]
    z_imag_list = [z.imag for z in z_list]
    z_inv_imag_list = [-z_imag for z_imag in z_imag_list]

    fig = plt.figure()
    plt.plot(z_real_list, z_inv_imag_list, 'o--', label=plot_label)

    plt.xlim(x_lim)
    plt.ylim(y_lim)
    plt.xlabel('Z$_{real}$ [$\Omega$]')
    plt.ylabel('-Z$_{imag}$ [$\Omega$]')
    plt.legend()
    plt.show()

def impedance_vs_fre_plot(fre_list, z_list, y_axis_stat='log10', x_axis_stat='log10'):
    """
    :param
        fre_list
        z_list
        x_axis_stat (Frequency)
            str
                'none':     normal
                'log10':    in log10
        y_axis_stat (Z, Z.real, Z.imag, Z.real + Z.imag)
            str
                'none':         Z normal
                'log10':        Z in log10
                'real':         Z.real in normal
                'real-log10':   Z.real in log10
                'imag':         Z.imag normal
                'imag-log10':   Z.imag in log10
    :return:
        x and y-axis are both in log(base=10) mode
    """
    fre_log10_list = [math.log(f, 10) for f in fre_list]

    z_norm_list = [math.sqrt(z.real ** 2 + z.imag ** 2) for z in z_list]
    z_norm_log10_list = [math.log(z, 10) for z in z_norm_list]

    if y_axis_stat == 'log10' and x_axis_stat == 'log10':
        fig = plt.figure()
        plt.plot(fre_log10_list, z_norm_log10_list, 'o--', label='Impedance - Frequency (log10 mode)')

        plt.xlabel('Log$_{10}$(Frequency) [Hz]')
        plt.ylabel('Log$_{10}$(Z) [$\Omega$]')
        plt.legend()
        plt.show()

def bode_two_plot(fre_list, z_list, fig_title='Bode-'):
    """
    Draw Impedance and Phase in two plot

    参考教程
        Creating multiple subplots using plt.subplots
        https://matplotlib.org/devdocs/gallery/subplots_axes_and_figures/subplots_demo.html
    """
    fre_log10_list = [math.log(f, 10) for f in fre_list]

    z_norm_list = [math.sqrt(z.real ** 2 + z.imag ** 2) for z in z_list]
    z_norm_log10_list = [math.log(z, 10) for z in z_norm_list]

    # tan_list = [z.imag / z.real for z in z_list] # z.real might be zero
    tan_list = []
    for z in z_list:
        if z.real == 0:
            tan_list.append(float('inf'))
        else:
            tan_list.append(z.imag / z.real)
    t_list = [math.atan(tan) * 180 / math.pi for tan in tan_list]

    fig, (ax1, ax2) = plt.subplots(2)
    fig.suptitle(fig_title)

    ax1.plot(fre_log10_list, z_norm_log10_list, 'o--', label='Impedance - Frequency (log10 mode)')

    ax1.set_xlabel(xlabel='Log$_{10}$(Frequency) [Hz]')
    ax1.set_ylabel(ylabel='Log$_{10}$(Z) [$\Omega$]')
    ax1.legend()

    ax2.plot(fre_log10_list, t_list, 'o--', label='Phase - Frequency')

    ax2.set_xlabel(xlabel='Log$_{10}$(Frequency) [Hz]')
    ax2.set_ylabel(ylabel='Phase (deg)')
    ax2.legend()

    plt.show()
```

```python
import datetime
import os
import pickle

"""
There are 10 ECMs now, creat configuration file for each ECM simulation
    1-Pickle parameters' configuration of each ECMs
    2-Load parameters' configuration of one ECM by ECM's serial
"""
# 1-Pickle parameters' configuration of each ECMs
def pickle_sim_ecm_para_config_dict(ecm_num, limit_list, para_list, fre_list, z_list):
    ecm_para_config_dict = {}
    ecm_para_config_dict['ecm_num'] = ecm_num
    ecm_para_config_dict['limit'] = limit_list
    ecm_para_config_dict['para'] = para_list
    ecm_para_config_dict['f'] = fre_list
    ecm_para_config_dict['z_sim'] = z_list

    # year, month, day int
    year = datetime.datetime.now().year
    month = datetime.datetime.now().month
    day = datetime.datetime.now().day

    file_name = str(year)+'_'+str(month)+'_'+str(day)+'_sim_ecm_'+str(ecm_num)+'_pickle.file'
    with open(file_name, 'wb') as file:
        pickle.dump(ecm_para_config_dict, file)
```

```python
def load_sim_ecm_para_config_dict(ecm_num, file_path):
    if ecm_num == 0:
        file_name = ''
    elif ecm_num == 1:
        file_name = ''
    elif ecm_num == 2:
        file_name = '2020_3_25_sim_ecm_2_pickle.file'
    elif ecm_num == 3:
        file_name = ''
    elif ecm_num == 4:
        file_name = ''
    elif ecm_num == 5:
        file_name = ''
    elif ecm_num == 6:
        file_name = ''
    elif ecm_num == 7:
        file_name = ''
    elif ecm_num == 8:
        file_name = ''
    elif ecm_num == 9:
        file_name = ''
    else:
        print('Currently, There only ECM- 0 ~ 9')

    file_abs_path = os.path.join(file_path, file_name)
    with open(file_abs_path, 'rb') as file:
        ecm_para_config_dict = pickle.load(file)
    return ecm_para_config_dict
```

### 尝试使用文献中的电路元件参数，绘制赖师兄的数据对应ECM-9

#### ECM-9-First Edition

First Edition: R0 = 0.007; Q0_pair= [q= 0.00015, n=0.9 ]; R1 = 70; Q1_pair = [q=0.0004, n = 0.7]; R2 = 30000;

不能拿ZSimView和Gamry软件进行模拟来对比结果
ZSimView高低频拟合结果很差，不能自主设置频率采样范围（没找到）,但是可以出入参数，给一个大致的图像趋势
Gamry设置的模型只能用于数据拟合，不能用于数据生成

```python
# ECM-9 R(Q(R(QR)))
def RaQaRaQRbbb(w, R0, Q0_pair, R1, Q1_pair, R2):
    z = R0 + 1 / ((1 / Q(w,q=Q0_pair[0],n=Q0_pair[1])) + (1/(R1 + (1/((Q(w, q=Q1_pair[0],n=Q1_pair[1])) + (1/R2))))))
    return z
```

```python
R0 = 0.007
Q0_pair= [0.00015, 0.9]
R1 = 70
Q1_pair = [0.0004, 0.7]
R2 = 30000

z_ecm_9_first = [RaQaRaQRbbb(w, R0, Q0_pair, R1, Q1_pair, R2) for w in w_list]
```

```python
nyquist_plot(z_list=z_ecm_9_first, x_lim=[0, 100], y_lim=[0, 60], plot_label='Nyquist-ECM9-Para(First Edition)')
```

上图只能看到一个时间常数，但根据自己的参数有两个，根据《导论》，一般两个时间常数相差5倍及以上能看到明显的不同圆弧（page78）
下图是上图的局部放大，在高频区的确出现一个感抗弧

```python
bode_two_plot(fre_list, z_list=z_ecm_9_first, fig_title='Bode-ECM9-Para(First Edition)')
```

### 尝试使用文献中的电路元件参数，绘制赖师兄的数据对应ECM-2

#### ECM-2-First Edition

First Edition: R0 = 0.015; Q0_pair= [q= 0.07, n=0.7 ]; R1 = 500; Q1_pair = [q=0.08, n = 0.8]; R2 = 400;
对应PyEIS的function cir_RsRQRQ

```python
# ECM-2 R(QR)(QR)
def RaQRbaQRb(w, R0, Q0_pair, R1, Q1_pair, R2):
    z = R0 \
        + 1 / ((1 / R1) + (1 / Q(w, q = Q0_pair[0], n = Q0_pair[1])))\
        + 1 / ((1 / R2) + (1 / Q(w, q = Q1_pair[0], n = Q1_pair[1])))
    return z
```

```python
R0 = 0.015
Q0_pair= [0.07, 0.7]
R1 = 500
Q1_pair = [0.08, 0.8]
R2 = 400

z_ecm_2_fir = [RaQRbaQRb(w, R0, Q0_pair, R1, Q1_pair, R2) for w in w_list]
```

```python
nyquist_plot(z_list=z_ecm_2_fir, x_lim=[0, 1000], y_lim=[0, 500], plot_label='Nyquist-ECM2-Para(First Edition)')
```

```python
bode_two_plot(fre_list, z_list=z_ecm_2_fir, fig_title='Bode-ECM2-Para(First Edition)')
```

```python
z_pyeis_ecm2 = pe.EIS_sim(frange = np.array(fre_list), \
                          circuit = pe.cir_RsRQRQ(w=np.array(w_list), \
                                                  Rs=R0, \
                                                  R=R1, \
                                                  n=Q0_pair[1], Q=Q0_pair[0], \
                                                  R2=R2, \
                                                  n2=Q1_pair[1], Q2=Q1_pair[0]), \
                          bode='log', legend='on')
```
